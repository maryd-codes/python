```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('data/countries.csv')
data.head(2)
```

# compare the populations of the 10 most populous countries in 2007

```python
data_2007 = data[data.year==2007]
data_2007.head(2)
```

```python
top10_2007 = data_2007.sort_values('population', ascending=False).head(10)
```

```python
x = range(10)# x = [0, 1, 2, ... 9]
plt.bar(x, top10_2007.population)
plt.show()
```

```python
plt.bar(x, top10_2007.population/10**6)
plt.xticks(x, top10_2007.country, rotation='vertical')
plt.title('10 most populous countries in 2007')
plt.ylabel('population in millions')
plt.show()
```

# Compare the GDP (NOT per capita!!) of the 10 most populous countries in 2007

```python
data_2007['gdpInBillions'] = (data_2007.population*data_2007.gdpPerCapita)/(10**9)
data_2007.head()
```

```python
top10_2007_byGDP = data_2007.sort_values('gdpInBillions', ascending=False).head(10)
top10_2007_byGDP
```
