```python
precip_sums_2017 = helper.precip_sums_for_year(year=2017)
precip_sums_2016 = helper.precip_sums_for_year(year=2016)
```

```python
# x = np.arange(len(helper.MONTHS)) # np.arange(12)
    
values_2018 = [month[1] for month in precip_sums_2018]
values_2017 = [month[1] for month in precip_sums_2017]
values_2016 = [month[1] for month in precip_sums_2016]
```

```python
# automatic stacked bar if you call on multiple data
plt.bar(x, values_2018)
plt.bar(x, values_2017)
plt.bar(x, values_2016)
plt.xticks(ticks=x, labels=helper.MONTHS);
```

```python
# making a grouped bar chart

# define variable for width of each group; width=1.0 = no space bx bars
# remember: first param of bar() is the POSITION of the bar!! Center default
group_width = 0.6
```

```python
plt.bar(x - group_width/3, values_2018, group_width/3) # "first bar": offset x-position to the left
plt.xticks(ticks=x, labels=helper.MONTHS);
```

```python
plt.bar(x - group_width/3, values_2018, group_width/3)
plt.bar(x, values_2017, group_width/3) # "second bar": no offset bc want it in the center
plt.bar(x + group_width/3, values_2016, group_width/3) # "third bar": offset x-position to the right
plt.xticks(ticks=x, labels=helper.MONTHS);
```

# Pie charts
plt.pie(data, 
        labels=[], 
        rotatelabels=True,
        labeldistance=1.5, --> labeldistance=1 = on the edge
        autopct='%.2f%%') --> floating point % with 2 dp

```python
wedges = np.array([8, 20, 20, 52])
plt.pie(wedges, 
        labels=['Green', 'Blue', 'Red', 'Blue'], 
        rotatelabels=True,
        labeldistance=1.5,
        autopct='%.2f%%');
```

```python
# small wedges can be offset: "exploding a wedge"

plt.pie(wedges, 
        labels=['Green', 'Blue', 'Red', 'Blue'], 
        rotatelabels=True,
        labeldistance=1.2,
        autopct='%.0f%%',
        explode=[0.5,0,0,0]);
```
