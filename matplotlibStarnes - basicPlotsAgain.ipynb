```python
data = np.random.randint(0,10,10)
x = np.arange(10)*10
plt.plot(x, data)
plt.title('Random data with some x-axis values specified')

plt.show()
```

```python
# get more control with xticks(tick locn, tick value)

import string
plt.plot(data)
plt.xticks(list(range(len(data))), list(string.ascii_uppercase[:len(data)]))

plt.show()

# N.B. xticks() doesn't check that iterables are the same length-- 
# if there are more labels than locations, it will disregard extras
# if there are more locations than labels, it will omit labels for the extra ticks.
```
