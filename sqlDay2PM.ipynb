```python
import sqlite3
import pandas as pd
```

```python
testdb = sqlite3.connect('data/test.db') # connect to the dummy database called test.db created yesterday
```

```python
pd.read_sql_query("SELECT * FROM orders;", testdb)
```

```python
cursor = testdb.cursor()
```

```python
cursor.execute('''CREATE TABLE employees
(empno INTEGER PRIMARY KEY,
first TEXT,
last TEXT,
role TEXT,
dateJoined DATE);
''')
testdb.commit()
# N.B. The commit function is invoked on the db object, not the cursor object.
```

```python
add_to_employees = [(1, "Joe", "Bloggs", "CEO", 2000-1-1),
           (2, "Sandra", "Bullock", "Sales Team Lead", 2000-1-1),
           (3, "Jane", "Doe", "Senior Sales Officer", 2000-1-1),
           (4, "Keanu", "Reaves", "Senior Sales Officer", 2000-1-1)]

cursor.executemany('''INSERT INTO employees
VALUES (?,?,?,?,?)
''', add_to_employees)
testdb.commit()
pd.read_sql_query("SELECT * FROM employees;", testdb)
```

```python
cursor.execute('''
SELECT role
FROM employees
INNER JOIN orders
ON orders.empno = employees.empno
WHERE price>400;''')
```

```python
pd.read_sql_query('''
SELECT employees.role, orders.price
FROM employees
INNER JOIN orders
ON orders.empno = employees.empno
WHERE price>400;''', testdb)
```
