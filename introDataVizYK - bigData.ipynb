```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Plotting aggregates of bigger data

```python
data = pd.read_csv('data/obama.csv', parse_dates=['year_month']) #parse_dates=[] makes timestamps
data.head(2)
```

```python
plt.plot(data.year_month, data.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

```python
approve_mean = data.groupby('year_month').mean()
approve_mean.head(2)
```

```python
plt.plot(approve_mean.index, approve_mean.approve_percent, 'red')

plt.plot(data.year_month, data.approve_percent, 'o', markersize=2, alpha=0.3)

plt.show()
```

# Plotting random sampling (of rows) of big data

```python
data_big = pd.read_csv('data/obama_too_big.csv', parse_dates=['year_month'])
data_big.head(2)
data_big.shape
```

```python
data.head(2)
data.shape
```

```python
plt.plot(data_big.year_month, data_big.approve_percent, 'o', markersize=2, alpha=0.3)

plt.show()
```

```python
sampled = data_big.sample(frac=0.1) # sample 10% of the data. Can use n=10000 to specify a number of samples instead

plt.plot(sampled.year_month, sampled.approve_percent, 'o', markersize=2, alpha=0.2)

plt.show()
```
